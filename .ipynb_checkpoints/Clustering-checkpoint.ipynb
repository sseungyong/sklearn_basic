```python
import os
from os.path import join
import copy
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import sklearn

import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
print(wine.DESCR)
```

```python
data = wine.data
label = wine.target
columns = wine.feature_names
```

```python
data = pd.DataFrame(data, columns = columns)
data.head()
```

```python
data.describe()
```

# Clustering

## k-Means clustering

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
pca_data.shape
```

#### model loading

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
```

#### fit data 

```python
kmeans.fit(pca_data)
```

```python
cluster = kmeans.predict(pca_data)
```

```python
cluster
```

```python
plt.scatter(pca_data[:,0], pca_data[:,1], c=cluster, linewidth=1, edgecolors='black')
plt.show()
```

## Hierarchical Clustering 

### single linkage : noise  

```python
from sklearn.cluster import AgglomerativeClustering
single_clustering = AgglomerativeClustering(n_clusters=3, linkage='single')
```

```python
single_clustering.fit(pca_data)
```

```python
single_cluster = single_clustering.labels_
single_cluster
```

```python
plt.scatter(pca_data[:,0], pca_data[:,1], c=single_cluster, linewidth=1, edgecolors='black')
plt.show()
```

##### 

```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(10,10))

# Hierachical Clustering  
children = single_clustering.children_

#         ,  .
distance = np.arange(children.shape[0])

#       .
no_of_observations = np.arange(2, children.shape[0]+2)

#     .
linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

#  .
dendrogram(linkage_matrix, p=len(pca_data), labels= single_cluster, show_contracted=True, no_labels=True)
plt.show()
```

### complete linkage : sigle  '' noise 

```python
complete_clustering = AgglomerativeClustering(n_clusters=3, linkage='complete')
```

```python
complete_clustering.fit(pca_data)
```

```python
complete_cluster = complete_clustering.labels_
complete_cluster
```

```python
plt.scatter(pca_data[:,0], pca_data[:,1], c=complete_cluster, linewidth=1, edgecolors='black')
plt.show()
```

#####  

```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(10,10))

# Hierachical Clustering  
children = complete_clustering.children_

#         ,  .
distance = np.arange(children.shape[0])

#       .
no_of_observations = np.arange(2, children.shape[0]+2)

#     .
linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

#  .
dendrogram(linkage_matrix, p=len(pca_data), labels= complete_cluster, show_contracted=True, no_labels=True)
plt.show()
```

###  average linkage

```python
average_clustering = AgglomerativeClustering(n_clusters=3, linkage='average')
```

```python
average_clustering.fit(pca_data)
```

```python
average_cluster = average_clustering.labels_
```

```python
plt.scatter(pca_data[:,0], pca_data[:,1], c=average_cluster, linewidth=1, edgecolors='black')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(10,10))

# Hierachical Clustering  
children = average_clustering.children_

#         ,  .
distance = np.arange(children.shape[0])

#       .
no_of_observations = np.arange(2, children.shape[0]+2)

#     .
linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

#  .
dendrogram(linkage_matrix, p=len(pca_data), labels= average_cluster, show_contracted=True, no_labels=True)
plt.show()
```

###  Compare

```python
plt.figure(figsize=(15,4))

plt.subplot(1, 3, 1)
plt.scatter(pca_data[:,0], pca_data[:,1], c=single_cluster)
plt.title('Sklearn Single Likage Hierarchical Clustering')

plt.subplot(1, 3, 2)
plt.scatter(pca_data[:,0], pca_data[:,1], c=complete_cluster)
plt.title('Sklearn Complete Likage Hierarchical Clustering')

plt.subplot(1, 3, 3)
plt.scatter(pca_data[:,0], pca_data[:,1], c=average_cluster)
plt.title('Sklearn Average Likage Hierarchical Clustering')
plt.show()
```

### Evaluation 

#### silhouette :
######    ,      . ,          . 

#### - k-means :   ,

```python
from sklearn.metrics import silhouette_score

best_n = 1
best_score = -1

for n_cluster in range(2, 11):
    kmeans = KMeans(n_clusters=n_cluster)
    kmeans.fit(pca_data)
    cluster = kmeans.predict(pca_data)
    score = silhouette_score(pca_data, cluster)
          
    print('num of cluster : {}, silhouette score : {:.2f}'.format(n_cluster, score))
    if score > best_score:
        best_score = score
        best_n = n_cluster
        
print('num of best score cluster : {}, best score : {}'.format(best_n, best_score))
```

#### - average linkage 

```python
from sklearn.metrics import silhouette_score

best_n = 1
best_score = -1

for n_cluster in range(2, 11):
    average_clustering = AgglomerativeClustering(n_clusters=n_cluster, linkage='average')
    average_clustering.fit(pca_data)
    cluster = average_clustering.labels_
    score = silhouette_score(pca_data, cluster)
          
    print('num of cluster : {}, silhouette score : {:.2f}'.format(n_cluster, score))
    if score > best_score:
        best_score = score
        best_n = n_cluster
        
print('num of best score cluster : {}, best score : {}'.format(best_n, best_score))
```
