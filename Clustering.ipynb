# Clustering

## k-Means clustering

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
pca_data.shape
```

#### model loading

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
```

#### fit data 

```python
kmeans.fit(pca_data)
```

```python
cluster = kmeans.predict(pca_data)
```

```python
cluster
```

```python
plt.scatter(pca_data[:,0], pca_data[:,1], c=cluster, linewidth=1, edgecolors='black')
plt.show()
```

## Hierarchical Clustering 

### single linkage : noise에 약함 

```python
from sklearn.cluster import AgglomerativeClustering
single_clustering = AgglomerativeClustering(n_clusters=3, linkage='single')
```

```python
single_clustering.fit(pca_data)
```

```python
single_cluster = single_clustering.labels_
single_cluster
```

```python
plt.scatter(pca_data[:,0], pca_data[:,1], c=single_cluster, linewidth=1, edgecolors='black')
plt.show()
```

##### 덴드로그램

```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(10,10))

# Hierachical Clustering의 자식 노드
children = single_clustering.children_

# 각 자식 노드간의 거리 정보를 가지고 있지 않기 때문에, 균일하게 그린다.
distance = np.arange(children.shape[0])

# 각 클러스터 단계를 포함한 노드의 수 계산한다.
no_of_observations = np.arange(2, children.shape[0]+2)

# 덴드로그램을 그리기위한 연결 매트릭스를 생성한다.
linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

# 덴드로그램을 그린다.
dendrogram(linkage_matrix, p=len(pca_data), labels= single_cluster, show_contracted=True, no_labels=True)
plt.show()
```

### Evaluation 

#### silhouette :
###### 같은 클러스터는 가까울 수록, 다른 클러스터와는 멀리 있을때 스코어가 높음. 단, 아웃라이어가 있는 경우는 예외적으로 높게 나오기도 함으로 전처리로 제거 필요. 

#### - k-means : 연산량이 비교적 높다,

```python
from sklearn.metrics import silhouette_score

best_n = 1
best_score = -1

for n_cluster in range(2, 11):
    kmeans = KMeans(n_clusters=n_cluster)
    kmeans.fit(pca_data)
    cluster = kmeans.predict(pca_data)
    score = silhouette_score(pca_data, cluster)
          
    print('num of cluster : {}, silhouette score : {:.2f}'.format(n_cluster, score))
    if score > best_score:
        best_score = score
        best_n = n_cluster
        
print('num of best score cluster : {}, best score : {}'.format(best_n, best_score))
```

#### - average linkage 

```python
from sklearn.metrics import silhouette_score

best_n = 1
best_score = -1

for n_cluster in range(2, 11):
    average_clustering = AgglomerativeClustering(n_clusters=n_cluster, linkage='average')
    average_clustering.fit(pca_data)
    cluster = average_clustering.labels_
    score = silhouette_score(pca_data, cluster)
          
    print('num of cluster : {}, silhouette score : {:.2f}'.format(n_cluster, score))
    if score > best_score:
        best_score = score
        best_n = n_cluster
        
print('num of best score cluster : {}, best score : {}'.format(best_n, best_score))
```
